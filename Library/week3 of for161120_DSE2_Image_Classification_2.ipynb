**สร้างฟังชั่น เพื่อใช้ scrape รูปจาก website 
/ สกัด feature ของแต่ละรูปออกมา / จัดเก็บในโฟลเดอร์**

```
def extractimagefeat(query):                                   # สร้างฟังชั่นชื่อ extractimagefeat มี parameter เป็น query
  #create folder
  foldername = 'images/'+query                                 # สร้าง folder เพื่อเก็บไฟล์ feature
  isdir = os.path.isdir(foldername)                            # os.path.isdir เป็น method ที่ใช้สำหรับตรวจสอบว่ามี foldername ใน directory แล้วหรือยัง (True/False)
  #check if folder exist
  if not isdir:                                                # ถ้ายังไม่มี ก็ให้สร้าง folder นี้                   
    #create directory
    os.makedirs(foldername)                                    # os.makedirs() เป็นคำสั่งในการสร้างโฟลเดอร์                         
  #get images from google search
  imglist = sc.get_image_urls(query)                           # สร้าง imglist เก็บภาพที่ scrape จาก google search       
  i = 1                      
  for img in imglist[0:30]:                                    # จากนั้นทำการวน loop ด้วยคำสั่ง for เพื่อทำวนทีละภาพ scrape มา 30 รูป >> [0:30] คือตำแน่งที่ 0-29
    #extract image features from each images and save to files
    try:                                                       # ใช้ try กับ except กรณีที่บางภาพไม่สามารถสกัด feature ได้ อาจจะด้วย format ของภาพที่ไม่รองรับ
      print(i)                                                 # ให้แสดงลำดับเลข ตอน execute 
      #create image path
      savepath = foldername + '/' + str(i)+'.p'                # สร้างไฟล์ path ของรูปแต่ละรูป  ตัวอย่าง >> image/หมา/25.p
      a = cv.image(img)                                        # ให้ a เป็น img ที่อยู่ใน imglist
      a.getresnet50()                                          # resnet เป็นโมเดลที่ใช้สำหรับสร้าง Features ทำการสร้าง feature จาก a                                       
      a.save(savepath)                                         # save เป็น file วน loop ไปจนครบทุกตัว
      i = i + 1
    except:
      pass
  return 'complete'                                            # ให้ return คำว่า complete 
```

สร้าง function เพื่อรวบรวมข้อมูลของ feature ทั้งหมด

เป็นการสร้าง list ของ dataset / สร้างตารางของ label และ feature

```
import glob                              # เป็น module ที่ใช้สำหรับหาไฟล์ที่ต้องการ
import pandas as pd                      # pandas เป็น module ที่ใช้จัดการกับไฟล์ ในที่นี้เราจะใช้สร้างตาราง
import pickle                            # module Pickle
def createdataset():                     # สร้าง function ชื่อ createdataset
  imgfolder = glob.glob('images/*')      # หาไฟล์ image/* มาเก็บใน imgfolder >> images/ข้าวมันไก่ images/ข้าวขาหมู
  #return imgfolder
  dataset = []                           # สร้าง list ชื่อ dataset
  for cls in imgfolder:                  # วน loop 
    clsset = pd.DataFrame()                # ให้ clsset สร้างตาราง pd.DataFrame() เป็นคำสั่งสำหรับสร้างตาราง
    pList = glob.glob(cls+'/*')            # ให้ pList สร้าง path >> images/ข้าวมันไก่/5.p
    #return pList
    featvec = []                           # สร้าง list ชื่อ featvec เพื่อเก็บ feature
    for p in pList:
      dat = pickle.load(open(p,'rb'))      # เปิดไฟล์ rb กำหนดสิทธิ์อ่านไฟล์ได้ Pickle เป็นไฟล์ที่สร้างไว้แล้วเราดึงกลับมาใช้ได้
      #return dat
      featvec.append(dat.resnet50)         # เพิ่ม feature ที่สกัดได้เข้าไปเก็บใน list ชื่อ featvec
    #return featvec
    #featvec = np.vstack(featvec)   
    clsset['feature'] = featvec            # ตาราง clsset ที่สร้างไว้ สร้าง column ชื่อ feature เก็บข้อมูล featvec
    cls = cls.split('/')[-1]               # สร้างชื่อ label โดยการ แยกชื่อ path ออกจากกัน และเอาเฉพาะตำแหน่งหลัง
    #return cls
    clsset['label'] = cls                  # ตาราง clsset ที่สร้างไว้ สร้าง column ชื่อ label เก็บข้อมูล ที่เป็นชื่อภาพ
    #return clsset
    dataset.append(clsset)                 # list dataset ที่สร้างไว้ ให้เพิ่มเข้าไปในตาราง
  #return dataset
  return pd.concat(dataset,axis=0)         # คืนค่า โดยเอา dataset ทั้งหมด มาต่อกันในแนวตั้ง axis = 0 >> pd.concat() เป็นการเอาข้อมูลมาต่อกัน
  
```

```
dataset = createdataset()
```

## Train model

```
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.svm import LinearSVC
#from sklearn.calibration import CalibratedClassifierCV
def trainmodel(dataset,modfile=''):
  trainfeat, testfeat, trainlabel, testlabel = train_test_split(dataset['feature'], dataset['label'], test_size=0.33, random_state=42)
  #return trainfeat,testfeat,trainlabel,testlabel
  clf = LinearSVC()
  #clf = CalibratedClassifierCV(clf) 
  mod = clf.fit(np.vstack(trainfeat.values),trainlabel.values)
  #return mod
  res = mod.predict(np.vstack(testfeat.values))
  #return res
  if modfile!='':
    pickle.dump(mod,open(modfile,'wb'))
  
  #return res, testlabel
  acc = sum(res == testlabel)/len(res)
  return mod,acc
```

```
mod,acc = trainmodel(dataset,'mymod.mod')
```

```
acc
```

```
# output function
modFile = 'mymod.mod'
mod = pickle.load(open(modFile,'rb'))
```

```
def predicting(imgurl):
  a = cv.image(imgurl)
  feat = a.getresnet50()
  res = mod.predict([feat])
  return res
```
